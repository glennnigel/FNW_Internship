```python
Rank=[]
State=[]
GSDP19_20=[]
GSDP18_19=[]
Share18_19=[]
GDP=[]



rnk='//td[@class="data1"]'
stat='//td[@class="name"]'
g19_20='//tr//td[3][@class="data"]'
g18_19='//tr//td[4][@class="data"]'
sh18_19='//tr//td[5][@class="data"]'
gd='//tr//td[6][@class="data"]'

rank=gdp.find_elements(By.XPATH,rnk)
for m in rank[1:34]:
    Rank.append(m.get_attribute('innerText'))

state=gdp.find_elements(By.XPATH,stat)
for m in state[:33]:
    State.append(m.get_attribute('innerText'))
    
gsdp1=gdp.find_elements(By.XPATH,g19_20)
for m in gsdp1[:33]:
    GSDP19_20.append(m.get_attribute('innerText'))
    
gsdp2=gdp.find_elements(By.XPATH,g18_19)
for m in gsdp2[:33]:
    GSDP18_19.append(m.get_attribute('innerText'))
    
share=gdp.find_elements(By.XPATH,sh18_19)
for m in share[:33]:
    Share18_19.append(m.get_attribute('innerText'))
    
gpd=gdp.find_elements(By.XPATH,gd)
for m in gpd[:33]:
    GDP.append(m.get_attribute('innerText'))
```
